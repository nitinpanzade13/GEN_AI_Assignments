### **Create a Simple AI-like Animation Script**

```python
%%manim -qm SimpleAnimation

class SimpleAnimation(Scene):
    def construct(self):
        text = Text("Hello Nitin 👋")
        self.play(Write(text))
        self.wait(2)

        circle = Circle()
        self.play(Transform(text, circle))
        self.wait(2)

        square = Square()
        self.play(Transform(circle, square))
        self.wait(2)
```

#### **Explanation:**
- %%manim → magic command to render video
- -qm → medium quality (fast rendering)
- Scene → main animation class
- self.play() → runs animation
- Transform() → changes one object to another

### **AI-Based Dynamic Animation Generator**

```python
def generate_animation(text_input):
    class AIGeneratedScene(Scene):
        def construct(self):
            text = Text(text_input)
            self.play(Write(text))
            self.wait(2)

            self.play(text.animate.scale(1.5))
            self.wait(1)

            self.play(FadeOut(text))

    return AIGeneratedScene
```

#### **Explanation:**
- This function acts like a mini AI tool
- Takes input text and generates animation dynamically
- You can pass any prompt → animation changes

### **Run AI Animation**

```python
scene = generate_animation("AI Animation 🚀")
%manim -qm scene
```

#### **Explanation:**
- Generates animation from user input
- Mimics AI-driven content creation
