```python
from transformers import pipeline
```

```python
qa_pipeline = pipeline(
    "question-answering",
    model="distilbert-base-cased-distilled-squad"
)

print("✅ QA model loaded successfully!")
```

```python
def qa_chatbot(context):
    print("🤖 QA Chatbot Started! (type 'exit' to stop)\n")
    print("📌 Context:\n", context)
    print("\n-------------------------\n")

    while True:
        question = input("You: ")

        if question.lower() == "exit":
            print("🤖 Bye brother 👋")
            break

        result = qa_pipeline(question=question, context=context)

        print("Bot:", result["answer"])
        print("Confidence:", round(result["score"], 3))
        print()
```

```python
context = """
Python is a programming language created by Guido van Rossum and released in 1991.
It is widely used for web development, artificial intelligence, data science, and automation.
Python is known for its simple syntax and readability.
"""

qa_chatbot(context)
```
