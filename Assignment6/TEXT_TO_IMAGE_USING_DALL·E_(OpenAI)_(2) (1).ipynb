Import Libraries

```python

# For displaying images in notebook
from IPython.display import Image, display

# PyTorch for GPU acceleration
import torch

# Stable Diffusion pipeline (local diffusion model)
from diffusers import StableDiffusionPipeline

from PIL import Image # For image processing

import time # For measuring inference time
```

Enable GPU Runtime

```python
# Detect GPU
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)
```

```python
# Load Stable Diffusion (optimized for 6GB RTX 3050)

pipe = StableDiffusionPipeline.from_pretrained(
    "runwayml/stable-diffusion-v1-5",
    torch_dtype=torch.float16
)

pipe = pipe.to(device)

# Memory optimizations
pipe.enable_attention_slicing()
pipe.enable_vae_slicing()
```

Generate Using Stable Diffusion

```python
prompt = "A futuristic AI classroom with robots teaching students"

start = time.time()

image = pipe(
    prompt,
    height=512,
    width=512,
    num_inference_steps=25,
    guidance_scale=7.5
).images[0]

end = time.time()

print("Generation Time:", round(end - start, 2), "seconds")

image.show()
```

## How Stable Diffusion Works

1. Text is encoded using CLIP text encoder.
2. Random noise is generated.
3. UNet gradually removes noise step-by-step.
4. VAE decodes latent representation into final image.

This process is known as latent diffusion.
