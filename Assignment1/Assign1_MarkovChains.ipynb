```python
import random
from collections import defaultdict
```

```python
class MarkovTextGenerator:
    def __init__(self):
        self.chain = defaultdict(list)

    def train(self, text):
        words = text.split()
        for i in range(len(words) - 1):
            self.chain[words[i]].append(words[i + 1])
        print("✅ Training completed!")

    def generate(self, start_word, length=30):
        if start_word not in self.chain:
            return "❌ Start word not found in training data."

        current = start_word
        output = [current]

        for _ in range(length - 1):
            next_words = self.chain.get(current)
            if not next_words:
                break
            current = random.choice(next_words)
            output.append(current)

        return " ".join(output)
```

```python
text_data = """
Machine learning is a field of artificial intelligence that gives computers the ability to learn.
Machine learning models are used in many applications such as recommendation systems and text generation.
Artificial intelligence is changing the world rapidly.
"""
```

```python
model = MarkovTextGenerator()
model.train(text_data)
```

```python
print(model.generate("Machine", length=30))
```
